```python
'''Question 3. Count records

How many taxi trips were totally made on January 15?

Tip: started and finished on 2019-01-15.

Remember that lpep_pickup_datetime and lpep_dropoff_datetime columns are in the format timestamp (date and hour+min+sec) and not in date.

    20689
    20530
    17630
    21090
'''

query = """
        SELECT
            count(1)
        FROM
            green_taxi_data
        WHERE
            date_trunc('day', lpep_pickup_datetime) = '2019-01-15'
            AND
            date_trunc('day', lpep_dropoff_datetime) = '2019-01-15'
        
"""
pd.read_sql(query, con=engine)

# A: 20530
```

```python
'''
Question 4. Largest trip for each day

Which was the day with the largest trip distance?
Use the pick up time for your calculations.

    2019-01-18
    2019-01-28
    2019-01-15
    2019-01-10
'''

query = """
        SELECT 
            date_trunc('day', lpep_pickup_datetime) 
        FROM 
            green_taxi_data
        WHERE 
            trip_distance = (
               SELECT MAX (trip_distance)
               FROM green_taxi_data)
        
"""
pd.read_sql(query, con=engine)

# A: 2019-01-15
```

```python
'''
Question 5. The number of passengers

In 2019-01-01 how many trips had 2 and 3 passengers?

    2: 1282 ; 3: 266
    2: 1532 ; 3: 126
    2: 1282 ; 3: 254
    2: 1282 ; 3: 274

'''

query = """
        SELECT 
            count(1)
        FROM 
            green_taxi_data
        WHERE 
            passenger_count = 2
            AND
            date_trunc('day', lpep_pickup_datetime) = '2019-01-01'
        
"""

query = """
        SELECT 
            count(1)
        FROM 
            green_taxi_data
        WHERE 
            passenger_count = 3
            AND
            date_trunc('day', lpep_pickup_datetime) = '2019-01-01'
        
"""
pd.read_sql(query, con=engine)

# A: 2: 1282, 3: 254
```

```python
'''
Question 6. Largest tip

For the passengers picked up in the Astoria Zone which was the drop off zone that had the largest tip? We want the name of the zone, not the id.

Note: it's not a typo, it's tip , not trip

    Central Park
    Jamaica
    South Ozone Park
    Long Island City/Queens Plaza
'''


query = ''' WITH my_query AS 
                (SELECT
                    tip_amount,
                    tzp."Zone" AS "PUZone",
                    tzd."Zone" AS "DOZone"
                FROM 
                    green_taxi_data t JOIN taxi_zone_lookup tzp
                    ON t."PULocationID" = tzp."LocationID"
                    JOIN taxi_zone_lookup tzd
                    ON t."DOLocationID" = tzd."LocationID"
                WHERE
                    tzp."Zone" LIKE 'Astoria')
            SELECT "DOZone"
            FROM my_query
            WHERE
                tip_amount = (SELECT MAX(tip_amount) FROM my_query)
'''
pd.read_sql(query, con=engine)

# A : Long Island City/Queens Plaza
```
